```python
url = 'https://api.bls.gov/publicAPI/v1/timeseries/data/'

key = '?registrationkey=d32f70d8f5d14c6ab8efc48d7296b131'

# Series stored as a dictionary
series_dict = {
    ' ENU04013105111150': 'county employment',
    'LAUCN281070000000003': 'local unemployment',
    'JTU110099000000000HIL': 'job opening',
    'APU000070111': 'average prices'
    }

# Start year and end year
dates = ('2020', '2023')
```

```python
# Specify json as content type to return
headers = {'Content-type': 'application/json'}

# Submit the list of series as data
data = json.dumps({
    "seriesid": list(series_dict.keys()),
    "startyear": dates[0],
    "endyear": dates[1]})

# Post request for the data
p = requests.post(
    '{}{}'.format(url, key),
    headers=headers,
    data=data).json()['Results']['series']
p
```
